```python
import urllib.request as req
from bs4 import BeautifulSoup
import csv

#indeed
LIMIT=50
INDEED=f"https://kr.indeed.com/%EC%B7%A8%EC%97%85?q=IT&limit={LIMIT}"

def extract_indeed_pages():
   
    result= req.urlopen(INDEED)
    soup = BeautifulSoup(result,"html.parser")
    pagination=soup.find("div",{"class":"pagination"})
    links= pagination.findAll("a")
    pages= []
    for link in links[:-1]:
        pages.append(int(link.string))
        max_page=pages[-1]
    return max_page

def extract_job(html):
    title=html.find("a",{"class":"jobtitle"})["title"]
    #company= html.find("span",{"class":"company"}).get_text(strip=True)
    loc=html.find("div",{"class":"recJobLoc"})["data-rc-loc"]
    #if company is None:
    #    company=None
    company=html.find("span",{"class":"company"})
    company_a = company.find("a")
    if company:
        if company_a is not None:
            company=str(company_a.string).strip()
        else: 
            company=str(company.string).strip()
    else:
        company=None
    job_id=html["data-jk"]
    return {'title':title,'company':company,'location':loc,
            "link":f"https://kr.indeed.com/%EC%B1%84%EC%9A%A9%EB%B3%B4%EA%B8%B0?jk={job_id}"}

def extract_indeed_jobs(last_page):
   
    jobs=[]
    for page in range(last_page): 
        print(f"scarapping indeed page {page}")
        result=req.urlopen(f"{INDEED}&start={page*LIMIT}")
        soup = BeautifulSoup(result,"html.parser")
        results= soup.findAll("div",{"data-tn-component":"organicJob"})
        for res in results:
            job=extract_job(res)
            jobs.append(job)
        
    return jobs

def get_indeed_jobs():
    last_page= extract_indeed_pages()
    jobs= extract_indeed_jobs(last_page)
    return jobs

#stack over flow
SOF=f"https://stackoverflow.com/jobs?q=python"

def extract_SOF_pages():
    result= req.urlopen(SOF)
    soup = BeautifulSoup(result,"html.parser")
    pagination=soup.find("div",{"class":"s-pagination"}).findAll("a",{"class":"s-pagination--item"})
    pages=[]
    for link in pagination[:-1]:
        pages.append(link.get_text(strip=True))
        max_page=pages[-1]
    return int(max_page)

def extract_job2(html):
    title= html.find("a",{"class":"s-link"})["title"]
    company=html.find("h3",{"class":"fc-black-700"}).find("span").get_text(strip=True)
    loc=html.find("span",{"class":"fc-black-500"}).get_text(strip=True)
    job_id =html["data-jobid"]
    return {"title":title,"company":company,"location":loc,"link":f"https://stackoverflow.com/jobs/{job_id}"}

def extract_SOF_jobs(last_page):
    jobs=[]
    for page in range(last_page):
        print(f"scarapping SOF page {page}")
        result=req.urlopen(f"{SOF}&pg={page+1}")
        soup = BeautifulSoup(result,"html.parser")
        results = soup.findAll("div",{"class":"-job"})
        for res in results:
            job= extract_job2(res)
            jobs.append(job)
    return jobs

def get_SOF_jobs():
    last_page=extract_SOF_pages()
    jobs=extract_SOF_jobs(last_page)
    return jobs

sof_jobs=get_SOF_jobs()
indeed_jobs=get_indeed_jobs()
jobs=indeed_jobs + sof_jobs

#save csv
def save_to_file(jobs):
    file= open("jobs.csv",encoding='utf-8-sig',mode="w")
    writer =csv.writer(file)
    writer.writerow(["title","company","location","link"])
    for job in jobs:
        writer.writerow(list(job.values()))
    print("done!")
    return

save_to_file(jobs)
```
